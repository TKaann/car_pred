```python
df = pd.read_excel("cars.xls")
```

```python
x = df[["Mileage","Cylinder","Liter","Cruise","Make","Trim"]]
```

```python
y = df[["Price"]]
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
x['Make'] = label_encoder.fit_transform(x['Make'])
x['Trim'] = label_encoder.fit_transform(x['Trim'])
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x_train,y_train)
```

```python
tahmin = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
(mean_squared_error(tahmin,y_test))**.5   #Label Encoding ile Fiyat farki
```

```python
r2_score(tahmin,y_test)    #Label Encoding Basari Orani
```

```python
x = df[["Mileage","Cylinder","Liter","Cruise","Make","Trim"]]
```

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x_train,y_train)
```

```python
tahmin = model.predict(x_test)
```

```python
(mean_squared_error(tahmin,y_test))**.5    #One-Hot Encoding ile Fiyat Farki
```

```python
r2_score(tahmin,y_test)      #One-Hot Dncoding ile Basari Orani
```
